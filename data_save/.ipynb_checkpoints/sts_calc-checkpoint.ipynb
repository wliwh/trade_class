### 涨跌统计

```python
CodeTimePd = {
    '000001':[2005,2007,2008,2009,2013,2015,2016,2018,2019,2021,2024],
    '399317':[2005,2008,2008,2010,2012,2015,2016,2017,2018,2021,2024],
    '000300':[2005,2007,2008,2009,2013,2015,2016,2018,2019,2021,2024],
    '000905':[2007,2008,2008,2010,2012,2015,2016,2017,2018,2021,2024],
    '000852':[2007,2008,2008,2010,2012,2015,2016,2016,2018,2021,2024],
    '399303':[2018,2022,2024]
}
```

```python
def split_index(code:str,ylst:list):
    ybeg = '{}-01-01'.format(ylst[0])
    yend = get_trade_day(16,date_fmt='%Y-%m-%d')
    p1 = funds_value.tdx_index_getter(code,ybeg,yend)
    lh_lst = list()
    ret_lst = list()
    for _i,yy in enumerate(ylst):
        ptmp = p1[p1.index.map(lambda x:x.startswith(str(yy)))]
        if _i%2==0:
            t1 = ptmp['close'].idxmin()
        else:
            t1 = ptmp['close'].idxmax()
        lh_lst.append(t1)
    for t1,t2 in zip(lh_lst[:-1], lh_lst[1:]):
        ret = p1.loc[t1:t2,'close']
        ret.index = pd.to_datetime(ret.index)
        ret_lst.append(ret)
    return ret_lst
```

```python
def plot_updown_line(a,b):
    adate, bdate = a.index[0].strftime('%Y-%m-%d'), b.index[0].strftime('%Y-%m-%d')
    aend, bend = a.index[-1].strftime('%Y-%m-%d'), b.index[-1].strftime('%Y-%m-%d')
    a1, b1 = a.to_list(), b.to_list()
    apct, bpct = 100*(a1[-1]/a1[0]-1), 100*(b1[-1]/b1[0]-1)
    if len(a1)>len(b1):
        b1 = b1 + [b1[-1] for _ in range(len(a1)-len(b1))]
    elif len(a1)<len(b1):
        a1 = a1 + [a1[-1] for _ in range(len(b1)-len(a1))]
    a1 = [a/a1[0] for a in a1]
    b1 = [b/b1[0] for b in b1]
    plt.plot(a1)
    plt.annotate(adate+';'+bdate,xy=(0,1))
    plt.annotate('%.2f %.2f' % (apct,bpct),xy=(int(len(a1)*0.8), 0.8 if a1[-1]<1 else 1.2))
    plt.plot(b1)
    plt.show()
```

```python
sz_lst = split_index('000300',CodeTimePd['000300'])
sm_lst = split_index('000905',CodeTimePd['000905'])
sms_lst = split_index('000852',CodeTimePd['000852'])
```
